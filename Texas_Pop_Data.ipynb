# Analysis of the 2018  Texas Senate Election

```python
table_nyt = pd.read_html('https://www.nytimes.com/interactive/2018/11/06/us/elections/results-texas-elections.html')
table_nyt_senate = pd.DataFrame(table_nyt[1])
table_nyt_senate.sort_values('County')
table_nyt_senate.head()
```

```python
#Let's sort the data by county [A-Z] so that we can easily look through the data. 

cln_nyt_senate = table_nyt_senate.dropna().sort_values('County')
cln_nyt_senate.head()
```

```python
#Let's add two new columns that show who won and by how much, this should make it easier to compare. 

cln_nyt_senate['Winner'] = np.where(cln_nyt_senate['Cruz']>cln_nyt_senate["O'Rourke"],'Cruz',"O'Rourke")
cln_nyt_senate['Dif'] = np.where(cln_nyt_senate['Winner'] == 'Cruz',cln_nyt_senate['Cruz'] - cln_nyt_senate["O'Rourke"],cln_nyt_senate["O'Rourke"] - cln_nyt_senate['Cruz'])
```
